```python
import numpy as np
import pandas as pd

```

```python
bad_match_table = pd.read_csv("tab_pix_to_dmd_coords.csv")
```

```python
bad_match_table
```

```python
new_xcentroids = bad_match_table['xcentroid'].values[::-1]
new_ycentroids = bad_match_table['ycentroid'].values[::-1]

new_good_match_table = bad_match_table.copy()

#new_good_match_table['xcentroid'] = new_xcentroids
new_good_match_table['ycentroid'] = new_ycentroids


new_good_match_table
```

```python
from astropy.wcs.utils import fit_wcs_from_points
from astropy.coordinates import SkyCoord
import astropy.units as u


dmd_skycoods = SkyCoord(new_good_match_table.x.values*u.arcsecond,
                        new_good_match_table.y.values*u.arcsecond,unit='deg')

ccdcoords = new_good_match_table[["xcentroid","ycentroid"]].values
ccd_to_dmd_wcs = fit_wcs_from_points(xy=ccdcoords.T,world_coords=dmd_skycoods, sip_degree=3)

mapfits = ccd_to_dmd_wcs.to_fits()

mapfits.writeto("flipped_DMD_Mapping_WCS.fits",overwrite=False)
```
